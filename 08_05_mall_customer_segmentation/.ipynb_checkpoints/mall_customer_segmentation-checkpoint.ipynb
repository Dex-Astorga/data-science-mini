```python
# IMPORT LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# LOAD DATASET
df = pd.read_csv('mall_customers.csv')
df.head(10)
```

```python
# RENAME MISPELLED COLUMNS
df.rename(columns={'Gnedere' : 'Gender'}, inplace=True)
df.columns
```

```python
# Examine the dataset
print('Data Shape:', df.shape)
df.describe()
```

```python
# CHECK DATA TYPES
df.dtypes
```

```python
# CHECK NULL VALUES
df.isnull().sum()
```

```python
# DROP IRRELEVANT COLUMNS
df.drop('CustomerID', axis=1, inplace=True)
```

```python
# BIN THE COLUMNS THEN VISUALIZE USING HISTPLOT
plt.figure(1, figsize=(15,4))   #1=figure1, 15=figureWidth, 6=figureHeight
n=0

for x in ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']: #traversing the 3-element list
    n+=1                                   #matplotlib indexing starts at 1, not 0, so we need to increment it right away
    plt.subplot(1, 3, n)                   #subplot(num_rows, num_columns, index(starting_at_position_1))
    plt.subplots_adjust(hspace=0.5, wspace=0.5) #adds styling
    sns.histplot(df[x], bins=20, color='#800040')           #seaborn histogram, formerly the distplot()
    sns.set_style('darkgrid')
    plt.title('Distplot of {}'.format(x))  #assign current value of x to the plot

plt.show()
```

```python
# GENDER DISTRIBUTION PLOT
monoc_colors = ['#952D61','#C32D61']  #creating custom pallete

plt.figure(figsize=(15,4))
sns.countplot(y='Gender', data=df, palette=monoc_colors) #countplot is ideal for visualizing categorical data
plt.show()
```

```python
# VIOLIN PLOT OF EACH COLUMN WITH RESPECT TO GENDER

plt.figure(2, figsize=(15,5)) #figure(figureNumber, figsize=(width, height))
n = 1                         #setting this to zero will make the firstplot different from the other plots because of the for loop structure

for col in ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']: #traversing 3-element list
    plt.subplot(1, 3, n)
    sns.set(style='darkgrid') #  'whitegrid', 'darkgrid', 'ticks', 'white', and 'dark'
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.violinplot(x=col, y='Gender', data=df, palette=monoc_colors)
    plt.ylabel('Gender Distribution' if n==1 else '')
    plt.title('Violin Plot {}'.format(col))
    n += 1

plt.show()
```

<p style="font-family: 'Georgia'; font-size: 14px; font-weight: 500; color: #800040;"> Violin plots are particularly useful when comparing the distribution of numerical variables across multiple categories. They provide a visual summary of the data's shape, central tendency, and spread within each category. By using violin plots, you can quickly identify differences and patterns in data distributions, aiding in data analysis and decision-making.</p>

```python
# VISUALIZING RELATIONHIP OF ANNUAL INCOME AND SPENDING SCORE USING SCATTERPLOT
sns.relplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, color='#800040', kind='scatter') #['scatter', 'line']
sns.set_style('darkgrid')
```

```python
# ASSIGNING AGE GROUPS
age_group1 = df.Age[(df.Age >= 18) & (df.Age <=25)]
age_group2 = df.Age[(df.Age >= 26) & (df.Age <=32)]
age_group3 = df.Age[(df.Age >= 33) & (df.Age <=40)]
age_group4 = df.Age[(df.Age >= 41) & (df.Age <=48)]
age_group5 = df.Age[(df.Age >= 49) & (df.Age <=55)]
age_group6 = df.Age[df.Age >= 55]

agex = ['18-25', '26-32', '33-40', '41-48', '49-55', 'Above 55']
agey = [len(age_group1.values), len(age_group2.values), len(age_group3.values), len(age_group4.values), len(age_group5.values), len(age_group6.values)] #.values is used to return a numpy array

plt.figure(3, figsize=(15,4))
sns.barplot(x=agex, y=agey, palette='rocket')
plt.title('Customer Ages')
plt.xlabel('Age')
plt.ylabel('Customer Count')
plt.show()
```

```python
# ASSIGNING SPENDING SCORE
ss_1_20 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 1) & (df['Spending Score (1-100)'] <= 20)]
ss_21_40 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 21) & (df['Spending Score (1-100)'] <= 40)]
ss_41_60 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 41) & (df['Spending Score (1-100)'] <= 60)]
ss_61_80 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 61) & (df['Spending Score (1-100)'] <= 80)]
ss_81_100 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 81) & (df['Spending Score (1-100)'] <= 100)]

ssx = ['ss_1_20', 'ss_21_40', 'ss_41_60', 'ss_61_80', 'ss_81_100']
ssy = [len(ss_1_20.values), len(ss_21_40.values), len(ss_41_60.values), len(ss_61_80.values),len(ss_81_100.values)]

plt.figure(3, figsize=(15,4))
sns.barplot(x=agex, y=agey, palette='rocket')
plt.title('Customer Spending Score')
plt.xlabel('Spending Score')
plt.ylabel('Customer Count')
plt.show()
```

```python
# ASSIGNING ANNUAL INCOME GROUPS
ai_0_30 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 0) & (df['Annual Income (k$)'] <= 30)]
ai_31_60 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 31) & (df['Annual Income (k$)'] <= 60)]
ai_61_90 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 61) & (df['Annual Income (k$)'] <= 90)]
ai_91_120 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 91) & (df['Annual Income (k$)'] <= 120)]
ai_121_150 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 121) & (df['Annual Income (k$)'] <= 150)]

aix = [ai_0_30, ai_31_60, ai_61_90, ai_91_120, ai_121_150]
aiy = [len(ai_0_30.values), len(ai_31_60.values), len(ai_61_90.values), len(ai_91_120.values),len(ai_121_150.values)]

plt.figure(3, figsize=(15,4))
sns.barplot(x=agex, y=agey, palette='rocket')
plt.title('Customer Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Customer Count')
plt.show()
```

```python
# FIND THE MOST OPTIMIZED NUMBER OF CLUSTERS BY TRYING 1-10
from sklearn.cluster import KMeans
import warnings                          #ignoring warnings
warnings.filterwarnings("ignore")

X1 = df.loc[:, ['Age', 'Spending Score (1-100)']].values   #create a new dataframe then find the 

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker='8')
plt.xlabel('k Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans.labels_, cmap='rainbow') 
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black')

plt.xlabel("Customer Clusters")
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.show()
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")

X2 = df.loc[:, ['Annual Income (k$)', 'Spending Score (1-100)']].values

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker='8')
plt.xlabel('k Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(X2)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X2[:, 0], X2[:, 1], c=kmeans.labels_, cmap='rainbow') 
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black')

plt.xlabel("Customer Clusters")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")

X3 = df.iloc[:, 1:]

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X3)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color='red', marker='8')
plt.xlabel('k Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6)
label = kmeans.fit_predict(X3)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
from mpl_toolkits.mplot3d import Axes3D

clusters = kmeans.fit_predict(X3)
df['label'] = clusters

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.label == 0], df['Annual Income (k$)'][df.label == 0], df['Spending Score (1-100)'][df.label == 0], c='red')
ax.scatter(df.Age[df.label == 1], df['Annual Income (k$)'][df.label == 1], df['Spending Score (1-100)'][df.label == 1], c='orange')
ax.scatter(df.Age[df.label == 2], df['Annual Income (k$)'][df.label == 2], df['Spending Score (1-100)'][df.label == 2], c='yellow')
ax.scatter(df.Age[df.label == 3], df['Annual Income (k$)'][df.label == 3], df['Spending Score (1-100)'][df.label == 3], c='green')
ax.scatter(df.Age[df.label == 4], df['Annual Income (k$)'][df.label == 4], df['Spending Score (1-100)'][df.label == 4], c='blue')
ax.scatter(df.Age[df.label == 5], df['Annual Income (k$)'][df.label == 5], df['Spending Score (1-100)'][df.label == 5], c='violet')
ax.view_init(30, 185)

plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
